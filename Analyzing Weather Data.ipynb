```python
import m5forecasting as m5
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wisconsin_weather = pd.read_csv("Weather/wisconsinw.csv")
```

```python
wisconsin_sales = m5.transformer[2](m5.all_data)[2]
sales_scaled = wisconsin_sales / wisconsin_sales.max()
```

```python
maxtemp = np.array(wisconsin_weather["maxtempC"].tolist())
maxtemp_scaled = (maxtemp - maxtemp.min()) / (maxtemp - maxtemp.min()).max()
```

```python
days = range(1400, 1700)
plt.figure(figsize=(15,6))
plt.plot(days, sales_scaled[days])
plt.plot(days, maxtemp_scaled[days])
plt.show()
```

```python
day = 1560 + 7 * 3
days = range(day, day + 28)
maxtemp = np.array(wisconsin_weather["maxtempC"].tolist())[days]
maxtemp_scaled = (maxtemp - maxtemp.min()) / (maxtemp - maxtemp.min()).max()
relevant_sales = wisconsin_sales[days]
sales_scaled = relevant_sales / relevant_sales.max()
print(wisconsin_weather["date_time"].iloc[days[0]])
print(wisconsin_weather["date_time"].iloc[days[-1]])
plt.figure(figsize=(15,6))
plt.plot(days, sales_scaled)
plt.plot(days, maxtemp_scaled)
plt.xticks(range(1, 2000, 7))
plt.xlim(days[0]- 3, days[-1] + 3)
plt.grid(1)
plt.show()
```
